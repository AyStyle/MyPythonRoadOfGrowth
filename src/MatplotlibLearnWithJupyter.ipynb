### 基础知识

#### 图形绘制

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 正弦波
x = np.linspace(0 , 2 * np.pi , 100)
y = np.sin(x)

# 开始画图
plt.plot(x,y)

# 设置图片的网格样式
plt.grid(linestyle = "--",color = "green",alpha = 0.65)

# 设置横纵坐标范围
plt.xlim([-1,10])
plt.ylim([-1.5,1.5])
```

#### 画布大小、坐标轴刻度、标签、标题

```python
# 画布大小，如果要设置画布大小，那么画布的设置一定要放在所有其他设置的前面
plt.figure(figsize = (9,6))

# 刻度设置
plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],[0,r"$\frac{\pi}{2}$",r"$\pi$",r"$\frac{3\pi}{2}$",r"$2\pi$"],fontsize=18)
plt.yticks([-1,-0.75,-0.5,0,0.5,0.75,1],fontsize = 18)

# 标签
plt.xlabel("x")
plt.ylabel("y = sin(x)",fontsize = 18)

# 标题
plt.title('正弦图像',fontsize = 24,color = "GREEN",fontfamily = "Kaiti")

plt.plot(x,y)



```

#### 图例

```python
x = np.arange(0 , 2*np.pi, step = np.pi/50)
plt.figure(figsize=(12,8))

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

# 添加图例，图例只能在线画出来之后添加
plt.legend(['Sin','Cos'],fontsize = 18)
```

#### 脊柱移动

```python
x = np.linspace(-np.pi,np.pi,100)
plt.figure(figsize=(12,8))
plt.plot(x,np.sin(x),x,np.cos(x))

# 获取轴面（get current axes）也称为子视图
axes = plt.gca()

axes.spines['top'].set_color('white')
axes.spines['right'].set_color('white')
axes.spines['right'].set_color('white')

# data 表示数据，表示数值
axes.spines['left'].set_position(('data',0))
axes.spines['bottom'].set_position(('data',0))

# 设置刻度
plt.yticks([-1,0,1])
_ = plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$\frac{-\pi}{2}$',0,r'$\frac{\pi}{2}$',r'$\pi$'],fontsize=18)

```

#### 图片保存

```python
 plt.figure(figsize=(12,8),linewidth=5)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.legend(['Sin','Cos'],loc = 'center',ncol = 2 ,bbox_to_anchor = (0,1.05,1,0.2),fontsize=18)

axes = plt.gca()
axes.set_facecolor("lightgreen")

# 保存图片
# dpi表示图片像素密度，越大越清晰
# bbox_inches = 'tight'，将图片设置为紧凑显示，保存整张图片
# facecolor表示背景色,edgecolor表示边框颜色
plt.savefig('./image.png',dpi = 100,bbox_inches = 'tight',facecolor='red',edgecolor='blue')
```

### 风格和样式

#### 颜色、线宽、线形、点形、透明度

```python
x = np.linspace(0,2*np.pi,20)
y1 = np.sin(x)
y2 = np.cos(x)

# linestyle线形设置
# linewidth线宽设置
# color颜色设置
# alpha透明度设置
# marker点形设置
# markersize点大小设置
plt.plot(x,y1,color = 'red',linestyle = '--',alpha = 0.3,marker = 'o')
plt.plot(x,y2,color = 'k', linestyle = '-.', alpha = 0.8,marker='*')
plt.plot(x,y1+y2,color = 'indigo',linestyle=':',marker='d')
plt.plot(x,y1-y2,color = '#FF00FF',linestyle=':',marker='1')
plt.plot(x,2*y1 + y2,color = (0.2,0.6,0.5),linestyle='-.',marker='3',markersize=20,linewidth = 5)
```

#### 更多设置

```python
x = np.linspace(0,5,50)
y = np.exp(-x) * np.cos(2 * np.pi * x)

plt.figure(figsize=(9,6))

plt.plot(x,y,marker= 'o',markersize=10,linewidth=3,color='GREEN',markerfacecolor='red',markeredgecolor='blue',markeredgewidth='2',alpha=0.8)
```

### 多图布局

#### 子视图

```python
plt.figure(figsize=(9,6))

x = np.linspace(0,2*np.pi,200)
y = np.sin(x)

# 子视图，子视图也称为轴面，左上角位置
axes = plt.subplot(2,2,1)
axes.set_facecolor('green')

axes.plot(x,y)

# 子视图，连起来写，右上角位置
axes = plt.subplot(222)
axes.set_facecolor('red')

axes.plot(x,y)

# 调用plot返回绘制对象，一条线
line, = axes.plot(x,y)
line.set_color('yellow')
line.set_linestyle('--')
line.set_marker('o')

axes = plt.subplot(2,1,2)
x = np.linspace(-np.pi,np.pi,200)
y = np.sin(x*x)
# axes.plot(x,y)

# 默认持有最近的子视图，索引调用plot方法，绘制图形就是向最近这个轴面绘制
plt.plot(x,y)
```

#### 嵌套

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

# fig就是整张图片
fig = plt.figure(figsize=(9,6))
plt.plot(x,y)

ax = fig.add_axes([0.2,0.2,0.2,0.2])
ax.plot(x, y, color = 'green', ls = '--')

plt.plot(x,y)

# 轴面[left, bottom, width, height]
ax = plt.axes([0.6,0.6,0.2,0.2])
ax.plot(x,y,color = 'red')

```

### 多图布局分隔显示

#### 均匀布局

```python
x = np.linspace(0,2*np.pi,200)

fig,((ax11,ax12,ax13),(ax21,ax22,ax23),(ax31,ax32,ax33)) = plt.subplots(3,3)
fig.set_figwidth(12)
fig.set_figheight(9)

ax11.plot(x,np.sin(x))
ax12.plot(x,np.cos(x))
ax13.plot(x,np.tan(x))

ax21.plot(x,np.tanh(x))
ax22.plot(x,np.sin(x*x))
ax23.plot(x,np.cos(x*x))

ax31.plot(x,np.sin(x) * np.cos(x))
ax32.plot(x,np.sin(x) + np.cos(x))
ax33.plot(x,np.sin(x*np.pi*x)*np.exp(-x))

# 设置紧凑布局
plt.tight_layout()
```

#### 不均匀布局

```python
from matplotlib import gridspec
```

```python
plt.figure(figsize = (12,9))

x = np.linspace(0,2*np.pi,200)

gs = gridspec.GridSpec(3,3)
ax = plt.subplot(gs[0,:]) # 默认情况下“:”沾满3列

ax.plot(x,np.sin(10*x))

ax = plt.subplot(gs[1,:2])
ax.plot(x,np.cos(x),color='red')
ax.set_facecolor("green")

ax = plt.subplot(gs[1:,2])
ax.plot(x,np.sin(x))

ax = plt.subplot(gs[2,0])
ax.plot(x,np.cos(2*np.pi*x)*np.exp(-x))

ax = plt.subplot(gs[2,1])
ax.plot([0,1,2],[0,1,2],marker='o')
```
